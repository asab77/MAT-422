```python
import matplotlib.pyplot as plt
import numpy as np

# Demonstrating Probability Axioms

# 1. Non-negativity
A_prob = 0.2  # Probability of Event A
B_prob = 0.5  # Probability of Event B

assert A_prob >= 0 and B_prob >= 0, "Probabilities must be non-negative!"

# 2. Normalization
total_probability = A_prob + B_prob
assert total_probability <= 1, "The total probability cannot exceed 1!"

# 3. Additivity - Mutually exclusive events
# If A and B are mutually exclusive, P(A ∩ B) = 0
P_A_or_B = A_prob + B_prob

# Plotting the probabilities of A and B
event = ['Event A', 'Event B']
probabilities = [A_prob, B_prob]

plt.bar(event, probabilities, color=['violet', 'green'])
plt.title('Probability of Events A and B')
plt.ylabel('Probability')
plt.show()
```

```python
# Conditional Probability Example
P_A_and_B = 0.2  # Probability of both A and B happening
P_B = 0.5  # Probability of Event B happening

# Conditional Probability of A given B
P_A_given_B = P_A_and_B / P_B

print(f"P(A | B) = {P_A_given_B:.2f}")
```

```python
# Discrete Random Variable Example - Rolling a die
die_outcomes = [1, 2, 3, 4, 5, 6]
probabilities = [1/6] * 6  # Each outcome has an equal probability of 1/6

plt.bar(die_outcomes, probabilities, color='blue')
plt.title('PMF of Rolling a Fair Die')
plt.xlabel('Outcome')
plt.ylabel('Probability')
plt.show()
```

```python
from scipy.stats import norm

# Continuous Random Variable Example - Normal Distribution
mu = 0  # Mean
sigma = 1  # Standard deviation

x = np.linspace(-4, 4, 1000)
pdf = norm.pdf(x, mu, sigma)

plt.plot(x, pdf, color='green')
plt.title('PDF of a Standard Normal Distribution')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.show()
```
