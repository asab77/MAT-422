# Necessary and Sufficient Conditions of Local Minimizers, Convexity, and Gradient Descent

## 3.3.1 Necessary and Sufficient Conditions of Local Minimizers

### First-order Necessary Condition:
A point \( x^* \) is a **critical point** if the gradient of the function \( f(x) \) vanishes:
\[
\nabla f(x^*) = 0
\]
This ensures that the slope of the function is zero at the critical point, indicating a potential local extremum (minimum or maximum).

### Second-order Sufficient Condition:
For a critical point \( x^* \) to be a **local minimum**, the Hessian matrix (second derivative) must be positive definite:
\[
\nabla^2 f(x^*) > 0
\]
This implies the function is curving upwards, confirming the point as a local minimum.

#### Example: \( f(x) = x^2 + 4x + 5 \)  
- **Gradient:** \( \nabla f(x) = 2x + 4 \)  
- **Hessian:** \( \nabla^2 f(x) = 2 \)  

Solving \( 2x + 4 = 0 \), we get:
\[
x^* = -2
\]
Since the Hessian at this point is \( 2 > 0 \), \( x = -2 \) is a **local minimum**.

---

## 3.3.2 Convexity and Global Minimizers

A function \( f(x) \) is **convex** if its second derivative is non-negative for all \( x \):
\[
\nabla^2 f(x) \geq 0, \quad \forall x
\]
Convex functions guarantee that any local minimum is also a **global minimum**.

#### Example:
- \( f_1(x) = x^2 \) is convex because \( \nabla^2 f_1(x) = 2 > 0 \) for all \( x \).
- \( f_2(x) = x^3 \) is non-convex because the second derivative changes sign: \( \nabla^2 f_2(x) = 6x \).

The plots illustrate:
- **Convex function:** \( f(x) = x^2 \) has a single global minimum.
- **Non-convex function:** \( f(x) = x^3 \) has regions where it decreases and increases, offering no global minimum guarantee.

## 3.3.3 Gradient Descent

### Gradient Descent Algorithm:
Gradient descent is an iterative optimization algorithm used to find the minimum of a function. Given a function \( f(x) \), the update rule is:
\[
x_{n+1} = x_n - \alpha \nabla f(x_n)
\]
where:
- \( x_n \) is the current point.
- \( \alpha \) is the learning rate, controlling the step size.
- \( \nabla f(x_n) \) is the gradient at the current point.

### Example: Minimizing \( f(x) = x^2 + 2x + 1 \)
- **Gradient:** \( \nabla f(x) = 2x + 2 \)
- **Update rule:** \( x_{n+1} = x_n - \alpha (2x_n + 2) \)

#### Execution with Learning Rate \( \alpha = 0.1 \):
Starting with \( x_0 = 5 \), the algorithm progresses through 10 iterations to converge near the minimum at \( x = -1 \). The step-by-step update is plotted, showing how \( x \) moves closer to the minimum with each iteration.

---

The visuals in the notebook highlight these core concepts:

1. **Local minimizers**: The critical point and Hessian confirm a minimum.
2. **Convexity**: We visually contrast a convex and non-convex function.
3. **Gradient descent**: The path of the descent is shown, demonstrating convergence to the minimum.


```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, diff, solve, hessian

# Section 3.3.1: Necessary and Sufficient Conditions of Local Minimizers
def local_minimizer():
    x = symbols('x')
    f = x**2 + 4*x + 5

    # First-order necessary condition: Gradient = 0
    grad_f = diff(f, x)
    critical_points = solve(grad_f, x)

    # Second-order sufficient condition: Hessian > 0
    hessian_f = diff(grad_f, x)

    print("Critical points:", critical_points)
    print("Hessian at critical point:", hessian_f.subs(x, critical_points[0]))

local_minimizer()

# Section 3.3.2: Convexity and Global Minimizers
def convexity_plot():
    x = np.linspace(-3, 3, 100)
    y1 = x**2  # Convex function
    y2 = x**3  # Non-convex function

    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    plt.plot(x, y1, label='$x^2$ (Convex)')
    plt.legend()
    plt.title("Convex Function")

    plt.subplot(1, 2, 2)
    plt.plot(x, y2, label='$x^3$ (Non-convex)', color='orange')
    plt.legend()
    plt.title("Non-Convex Function")

    plt.show()

convexity_plot()

# Section 3.3.3: Gradient Descent
def gradient_descent(learning_rate=0.1, iterations=10):
    x = 5  # Initial guess
    history = [x]

    for _ in range(iterations):
        grad = 2 * x + 2  # Gradient of f(x) = x^2 + 2x + 1
        x = x - learning_rate * grad
        history.append(x)

    print("Minimum found at:", x)

    # Plotting the gradient descent path
    plt.plot(history, marker='o')
    plt.title('Gradient Descent Progress')
    plt.xlabel('Iteration')
    plt.ylabel('x value')
    plt.show()

gradient_descent()
```
