## 2.3.1 Joint Probability Distributions

In probability theory, a joint probability distribution shows the probability of two events happening simultaneously. If we have two discrete random variables $ X $ and $ Y $, their joint probability is denoted as $ P(X = x, Y = y) $.\\

Mathematically:

$P(X = x, Y = y) = P(X = x) \times P(Y = y) \quad \text{if X and Y are independent}$


```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Example of joint probability for two random variables
x = np.random.choice([0, 1], size=1000, p=[0.5, 0.5])
y = np.random.choice([0, 1], size=1000, p=[0.7, 0.3])

# Create a dataframe for joint distribution
joint_df = pd.crosstab(x, y, rownames=['X'], colnames=['Y'], normalize=True)

# Visualize the joint probability distribution
sns.heatmap(joint_df, annot=True, cmap='Blues')
plt.title("Joint Probability Distribution of X and Y")
plt.show()
```

## 2.3.2 Correlation and Dependence

Correlation measures the strength and direction of a linear relationship between two variables. The Pearson correlation coefficient $r$ is defined as:

$$r = \frac{\text{cov}(X, Y)}{\sigma_X \sigma_Y}$$
Where:
- $\text{cov}(X, Y)$ is the covariance of $X$ and $Y$
- $\sigma_X$ and $\sigma_Y$ are the standard deviations of $X$ and


```python
# Generate two random variables with some correlation
np.random.seed(0)
x = np.random.rand(100)
y = 2 * x + np.random.rand(100)

# Calculate Pearson correlation coefficient
correlation = np.corrcoef(x, y)[0, 1]
print(f"Pearson Correlation Coefficient: {correlation}")

# Scatter plot to visualize correlation
plt.scatter(x, y)
plt.title(f"Scatter Plot of X and Y (Correlation: {correlation:.2f})")
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

## 2.3.3 Random Samples

A random sample refers to a subset of data drawn from a larger population, where each data point is selected randomly. This is used in statistics to make inferences about the population.

The law of large numbers states that as the sample size increases, the sample mean will converge to the population mean.


```python
# Generating random samples from a normal distribution
mean, std_dev = 0, 1  # mean and standard deviation
sample = np.random.normal(mean, std_dev, 1000)

# Calculate sample mean and compare with population mean
sample_mean = np.mean(sample)
print(f"Sample Mean: {sample_mean}")

# Visualize the sample distribution
plt.hist(sample, bins=30, density=True, alpha=0.6, color='g')
plt.title("Histogram of Random Samples from Normal Distribution")
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```
